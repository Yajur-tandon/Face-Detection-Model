```python
import Augmentor
import os

P = Augmentor.Pipeline('Yajur Tandon')
P.random_brightness(probability=0.7, min_factor=0.3, max_factor=0.8)
P.zoom_random(probability=0.5, percentage_area=0.8)
P.flip_left_right(probability=0.8)
P.sample(100)

```

```python

def face_augmentor(root_dir: str):
    files = os.listdir(root_dir)
    for f in files:
        full_path = os.path.join(root_dir, f)
        P = Augmentor.Pipeline(full_path)
        P.random_brightness(probability=0.7, min_factor=0.3, max_factor=0.8)
        P.zoom_random(probability=0.5, percentage_area=0.8)
        P.flip_left_right(probability=0.8)
        P.sample(100)

root_dir = 'face_data_2'
face_augmentor(root_dir=root_dir)
```
